```python
data = np.genfromtxt(r"C:\Users\hp\Downloads\0000000000002417_training_boston_x_y_train.csv",delimiter=",")
print(data.shape)
df=pd.DataFrame(data)
y=data[:,13]
x=data[:,0:13]
x=np.delete(data,13,1)
c=np.ones((379,1))
#scaler.fit(x)
#x=scaler.transform(x)
x=np.append(x,c,1)
```

```python
def cost(points,m):
    total_cost=0
    M=len(x)
    N=len(x[0])
    for i in range(M):
        X=x[i]
        Y=y[i]
        total_cost+=(1/M)*((Y-(m*X).sum())**2)
    return total_cost
```

```python
def step_gradient(points,lr,m):
    N=len(x[0])
    m_slope=np.zeros(N)
    M=len(x)
    for i in range(N):
        for j in range(M):
            m_slope[i] += (-2/M)*(y[j]-(m*x[j,0:N]).sum())*x[j,i]
    new_m=m-lr*m_slope
    return new_m
```

```python
def gd(points,lr,num):
    N=len(x[0])   
    m=np.zeros(N)
    M=len(x)
    for i in range(num):
        m=step_gradient(points,lr,m)
        print(i,"  cost:",cost(points,m))
    return m
```

```python
def run():
    data=np.genfromtxt(r"C:\Users\hp\Downloads\0000000000002417_training_boston_x_y_train.csv",delimiter=",")
    learning_rate=0.01
    num_iterations=1000
    m = gd(data,learning_rate,num_iterations)
    print(m)
    return m
```

```python
m=run()
```
